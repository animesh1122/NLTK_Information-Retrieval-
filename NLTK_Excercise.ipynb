# Extract & parsing the text data from "https://en.wikipedia.org/wiki/COVID-19_pandemic_in_New_York_(state)"

```python
import nltk
from urllib import request
url = "https://en.wikipedia.org/wiki/COVID-19_pandemic_in_New_York_(state)"
html = request.urlopen(url).read().decode('utf8')
```

```python
from bs4 import BeautifulSoup
raw = BeautifulSoup(html, 'html.parser').get_text()
raw
```

# Apply word tokenization to the file

```python
from nltk import word_tokenize
tokens = word_tokenize(raw)
```

```python
type(tokens)
```

```python
tokens[:150]
```

#  Normalize all tokens into lowercase & apply wordnet Lemmatization to the tokens.

```python
tokens = word_tokenize(raw)
type(tokens)
```

```python
words = [w.lower() for w in tokens]
words
```

```python
wnl = nltk.WordNetLemmatizer()
wnl_tokens = [wnl.lemmatize(t) for t in tokens]
wnl_tokens
```

# Generate vocabulary (distinct words after normalization & lemmatization), and write the vocabulary to the text file "vocab.txt". 

```python
 vocab = sorted(set(words))
```
